```python
import dateutil.relativedelta
from datetime import datetime
```

```python
to_date=datetime.now()
to_date=datetime.strftime(to_date,'%Y,%m,%d')
to_date=datetime.strptime(to_date,'%Y,%m,%d')
from_date=to_date-dateutil.relativedelta.relativedelta(month=1)
```

```python
train["rollin_window10"]=train.Close.rolling(10, min_periods = 1).mean()
```

```python
train["pre_volume"]=train["Volume"].shift(periods=1)
```

```python
train.fillna(0,inplace=True)
```

```python
train.head()
```

```python
train["volume_shocks"]=pd.Series([((train["Volume"]-train["pre_volume"]/train["Volume"])*100>10) | ((train["Volume"]-train["pre_volume"]/train["Volume"])*100<10)])
```

```python
train["volume_shocks"]=train["volume_shocks"].astype("bool")
```

```python
train.head()
```

```python
train["price shock"]=(((train["Close"].shift(periods=1)-train["Close"])*100/train["Close"])>2).astype("Bool")
```

```python
train["black_swan"]=(((train["Close"].shift(periods=1)-train["Close"])*100/train["Close"])>2).astype("Bool")
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
from math import log
```

```python
def test_stationarity(x):
    
    #Determing rolling statistics
    rolmean = x.rolling(12).mean()
    rolstd = x.rolling(12).std()

    #Plot rolling statistics:
    orig = plt.plot(x, color='blue',label='Original')
    mean = plt.plot(rolmean, color='red', label='Rolling Mean')
    std = plt.plot(rolstd, color='black', label = 'Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show(block=False)
```

```python
test_stationarity(train["Close"])
```
